## MATPLOTLIB

### Raw 3D Model

```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

num_points = 100 # random num points
x = np.random.randint(0, 100, size=num_points)
y = np.random.randint(0, 100, size=num_points)
z = np.random.randint(0, 100, size=num_points)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

#create plot
ax.scatter(x, y, z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

### K-Means Clustered Model

```python
from sklearn.cluster import KMeans

data = np.vstack((x, y, z)).T

num_clusters = 3 #random num clusters
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(data)
labels = kmeans.labels_

scatter = ax.scatter(x, y, z, c=labels)

legend1 = ax.legend(*scatter.legend_elements(), title="clusters")
ax.add_artist(legend1)

plt.show()
```

## MAYAVI

### Raw 3D Model

```python
from mayavi import mlab

x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)

fig = mlab.figure()

mlab.points3d(x, y, z, scale_factor=0.1)

mlab.xlabel('x')
mlab.ylabel('y')
mlab.zlabel('z')

mlab.show()
```

### K-Means Clustered Model

```python
points = np.column_stack((x, y, z))

kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(points)

fig = mlab.figure()

num_clusters = 3
for cluster_num in range(num_clusters):
    cluster_points = points[cluster_labels == cluster_num]
    mlab.points3d(cluster_points[:, 0], cluster_points[:, 1], cluster_points[:, 2],
                  scale_factor=0.1, color=tuple(np.random.rand(3)))

mlab.xlabel('x')
mlab.ylabel('y')
mlab.zlabel('z')

mlab.show()
```
